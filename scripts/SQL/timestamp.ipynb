```python
import pandas as pd
```

```python
df = pd.read_csv('preprocessed_hotel 수정.csv')
```

```python
df.drop('date_pair', axis=1, inplace=True)
```

```python
df
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
```

```python
df['start_date'] = df['start_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
df['end_date'] = df['end_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df.to_csv('preprocessed_hotel 수정.csv', index=False)
```
